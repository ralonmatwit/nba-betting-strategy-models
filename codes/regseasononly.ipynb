```python
import pandas as pd

# Load games dataset
games_df = pd.read_csv("nba_games_all.csv", low_memory=False)

# Convert game_date to datetime
games_df["game_date"] = pd.to_datetime(games_df["game_date"], errors="coerce")

# Drop rows with missing dates
games_df = games_df[games_df["game_date"].notna()]

# Keep only Regular Season games
games_df = games_df[games_df["season_type"] == "Regular Season"]

# Reset index
games_df = games_df.reset_index(drop=True)

# Export cleaned dataset
games_df.to_csv("nba_games_regular_seasons.csv", index=False)

print("Final dataset shape:", games_df.shape)
print(games_df[["game_id","game_date","matchup","season","season_type"]].head())
```

```python
import pandas as pd

# Load the cleaned dataset
games_df = pd.read_csv("nba_games_2010_onward.csv", low_memory=False)

# Ensure game_date is datetime
games_df["game_date"] = pd.to_datetime(games_df["game_date"], errors="coerce")

# Drop rows with missing dates (just in case)
games_df = games_df[games_df["game_date"].notna()]

# Sort by date ascending (oldest â†’ newest)
games_df = games_df.sort_values(by="game_date").reset_index(drop=True)

# Export organized dataset
games_df.to_csv("nba_games_2010_onward_sorted.csv", index=False)

print("Final dataset shape:", games_df.shape)
print(games_df[["game_id","game_date","matchup","team_id","pts"]].head())
print(games_df[["game_id","game_date","matchup","team_id","pts"]].tail())
```
